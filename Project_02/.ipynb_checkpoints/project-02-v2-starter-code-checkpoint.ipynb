**1: Describe how the covariance matrix is different from the correlation matrix.**

Although covariance and correlation both measures how variables change with each other, covariance does not provide a strength of relationship due to different scaling of variables. Correlation overcomes this problem by normalising covariance to provide a means of comparison across variables.


**2: What is the process to convert the covariance into the correlation?**

To convert covariance to correlation, we normalise covariance by dividing it with the std deviations of individual variables.

**3: Why is the correlation matrix preferred to the covariance matrix for examining relationships in your data?**

Correlation provides a mean of comparison on the strength of linear relationship between variables but covariance does not.

** Question and deliverables**

To test if alcohol use remains the same after age of 30 at alpha of 5%

*H0*: Alcohol usage is the same before and after age of 30

*H1*: Alcohol usage is **not** the same before and after age of 30


````python
before = drug_use.loc[:12, ['age', 'alcohol-use']]
after = drug_use.loc[13:, ['age', 'alcohol-use']]
````

````python
# Bar chart of alcohol use before and after age of 30

beforeplot = before.plot.bar()
beforeplot.set_title('Bar: Age versus Alcohol use')
beforeplot.set_xticklabels(drug_use['age'], rotation=45 ) 

afterplot = after.plot.bar()
afterplot.set_title('Bar: Age versus Alcohol use')
afterplot.set_xticklabels(drug_use['age'], rotation=45 ) 
````

````python
result = stats.ttest_ind(after['alcohol-use'], before['alcohol-use'])

pvalue = result[1]

print ('P-value is ' + str(pvalue))
````

**Report**

Since the p-value is 33% which is much greater than our threshold of 5%, we cannot reject the null hypothesis. In this case, there is no difference between alcohol consumption before and after age of 30.

Outliers are an interesting problem in statistics, in that there is not an agreed upon best way to define them. Subjectivity in selecting and analyzing data is a problem that will recur throughout the course.

1. Pull out the rate variable from the sat dataset.
2. Are there outliers in the dataset? Define, in words, how you _numerically define outliers._
3. Print out the outliers in the dataset.
4. Remove the outliers from the dataset.
5. Compare the mean, median, and standard deviation of the "cleaned" data without outliers to the original. What is different about them and why?

````python
sat_df1.describe()
````

````python
# C
iqr = 63.5 - 9

outlier_upper = 1.5*iqr + 63.5
outlier_lower = 9 - 1.5*iqr

print('Outlier band: ', (outlier_lower, outlier_upper))
````

#### My Assumptions

Since the range of Rate is within the outlier band, there seem to be no outliers in Rate.

However, participation rates across states are influenced by different policies and education systems.

From this angle when analysing the data, it may make more sense to have comparisons across states with similar participation rates within a tolerable range of variation against scores.

For this exercise, we will assume 1 std deviation from the mean will be tolerable for the purpose of comparability across states.

````python
# Finding the upperlimit and lowerlimit to determine outliers
upperlimit = sat_df1['Rate'].mean() + sat_df1['Rate'].std()
lowerlimit = sat_df1['Rate'].mean() - sat_df1['Rate'].std()

# Identifying rows where rate is above upperlimit or below lowerlimit
outliers = sat_df1[(sat_df1['Rate'] > upperlimit) | (sat_df1['Rate'] < lowerlimit)]
````

````python
# Print table of outliers
outliers
````

````python
# Removing outliers from sat_df1 and assign result to new dataframe sat_cleaned

sat_cleaned = sat_df1.drop(outliers.index)
````

````python
sat_cleaned.head()
````

````python
sat_df1['Rate'].describe()
````

````python
sat_cleaned['Rate'].describe()
````

#### Comparison of mean, median and std deviation before and after removal of outliers

With the removal of the outliers, we have observed that mean has dropped slightly. This would indicate that they may be more outliers above the upperlimit vs below the lowerlimit.

Median is observed to have increased slightly, which further suggests that they may be more outliers above the upperlimit vs below the lowerlimit.

The most observable difference is from std deviation where it dropped from 27.6 to 19.3 (30% reduction). This is expected with the removal of outliers, where tail cases are removed thus overall variation will be reduced significantly.

### 9. Percentile scoring and spearman rank correlation

---

### 9.1 Calculate the spearman correlation of sat `Verbal` and `Math`

1. How does the spearman correlation compare to the pearson correlation? 
2. Describe clearly in words the process of calculating the spearman rank correlation.
  - Hint: the word "rank" is in the name of the process for a reason!


````python
# Spearman correlation

sat_df1[['Verbal', 'Math']].corr(method='spearman')
````

````python
# Pearson correlation

sat_df1[['Verbal', 'Math']].corr()
````

The spearman correlation is slightly higher than that of pearson correlation. This means the monotonic relationship (directional relationship) measured by spearman is slightly stronger than that of the lineaer relationship as measured by pearson.

To calculate the spearman correlation:
**(Note: do not change the order or sort the values of the variable)**
1. For each variable, first assign a rank to individual values according to their magnitude. For eg, the value with the biggest magnitude (i.e. max value of the variable) will be assigned a rank of 1, the second largest value will be assigned rank 2, and the Nth largest value will be assign N. 
<br><br>
2. If there are ties in rank, the average of the ranks which those values would have occupied will be assigned instead. To illustrate, suppose there are 2 occurences of 55 in a variable and they would have been assigned rank 10 and 11 respectively. However in this case, since there is a tie we will assign 10.5 to both values instead (average of 10 and 11). Rank 10 and 11 will then be skipped.
<br><br>

### 9.3 Percentiles and outliers

1. Why might percentile scoring be useful for dealing with outliers?
2. Plot the distribution of a variable of your choice from the drug use dataset.
3. Plot the same variable but percentile scored.
4. Describe the effect, visually, of coverting raw scores to percentile.

#### Answer
Percentile scoring allows us to quickly identify the 25th and 75th percentile for the calculation of interquartile range which is useful for dealing with outliers.

````python
# Distribution plot for Alcohol Use

fig = plt.figure(figsize=(7,7))

alcohol = sns.distplot(drug_use['alcohol-use'], bins=10,kde=False)

alcohol.set_title('Alcohol Usage Distribution', fontsize=20)

alcohol.set_ylabel('Frequency', fontsize=16)

alcohol.set_xlabel('Alcohol use', fontsize=16)
````

````python
# Percentile scoring Alcohol Use column and assign results to variable alcohol_percentile

alcohol_percentile = [stats.percentileofscore(drug_use['alcohol-use'],use) for use in drug_use['alcohol-use']]
````

````python
# Distribution plot for percentile of Alcohol Use

fig = plt.figure(figsize=(7,7))

percentile = sns.distplot(alcohol_percentile, bins=10, kde=False)

percentile.set_title('Alcohol Usage Percentile Distribution', fontsize=20)

percentile.set_ylabel('Frequency', fontsize=16)

percentile.set_xlabel('Percentile', fontsize=16)
````

By converting raw scores to percentile, we have changed the shape of the distribution to be relatively uniformed.
